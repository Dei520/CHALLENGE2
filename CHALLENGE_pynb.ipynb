```python
import pandas as pd


data = {
    "ID": [1, 2, 3, 4, 5],
    "Nombre": ["Juan", "Ana", "Pedro", "Mar칤a", "Luis"],
    "Edad": [23, 45, 34, 28, 40],
    "Ciudad": ["CDMX", "Guadalajara", "Monterrey", "CDMX", "Puebla"],
    "Evasion": [1, 0, 1, 0, 0]  # 1 = evasi칩n, 0 = sin evasi칩n
}


df = pd.DataFrame(data)


print("游댌 Columnas y tipos de datos:")
print(df.dtypes)
print("\nPrimeras filas del dataset:")
print(df.head())


diccionario = {
    "ID": "Identificador 칰nico de cada registro",
    "Nombre": "Nombre de la persona",
    "Edad": "Edad en a침os",
    "Ciudad": "Ciudad de residencia",
    "Evasion": "Indica si hubo evasi칩n (1 = s칤, 0 = no)"
}

print("\n游닀 Diccionario de variables:")
for columna, descripcion in diccionario.items():
    print(f"{columna}: {descripcion}")


columnas_relevantes = ["Edad", "Ciudad", "Evasion"]
df_relevante = df[columnas_relevantes]

print("\n游늵 Columnas m치s relevantes para el an치lisis de evasi칩n:")
print(df_relevante)
```
